# OEMC FAIR Checker

This notebook:
1. Fetches a GeoKnowledge Hub package and summarises it
2. Parses and displays all knowledge resources (datasets, publications, software, other)
3. Assesses FAIR compliance **only for datasets** that have Zenodo DOIs
4. Prints a summary report

```python
import requests
from fair_checker import (
    GKHubClient,
    summarize_package,
    parse_knowledge_resources,
    pretty_print_knowledge_resources,
    assess_all_knowledge_resources_fair,
    print_summary_report,
    assess_fair_compliance,
    print_fair_report,
    write_report,
)
```

## 1. Fetch & summarise GKHub package

```python
PACKAGE_ID = '4eta5-f1z60' # GKHUB Use case URL: https://gkhub.earthobservations.org/packages/4eta5-f1z60

client = GKHubClient()
pkg = client.get_package(PACKAGE_ID)
summarize_package(pkg)
```

## 2. Parse & display knowledge resources

```python
BASE = "https://gkhub.earthobservations.org"
html_text = requests.get(f"{BASE}/packages/{PACKAGE_ID}", timeout=30).text
resources = parse_knowledge_resources(html_text, PACKAGE_ID)

print(f"Found: {len(resources['datasets'])} datasets, "
      f"{len(resources['journal_publications'])} publications, "
      f"{len(resources['software'])} software, "
      f"{len(resources['other'])} other")

pretty_print_knowledge_resources(resources, title=pkg.get("title"), package_id=PACKAGE_ID)
```

## 3. Assess FAIR compliance for all datasets

```python
results = assess_all_knowledge_resources_fair(resources)
print_summary_report(results)
```

## 4. Export report to file

```python
report_name = f"FAIR_assessment_{PACKAGE_ID}.md"
write_report(pkg, resources, results, report_name)
# or specify a path: write_report(pkg, resources, results, "my_report.md")
```
